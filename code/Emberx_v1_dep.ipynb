```python
import torch
from ultralytics import YOLO

# Check for GPU and set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load the model
model_path = '/content/emberx_v1.pt'
try:
    # Use ultralytics.YOLO to load the model
    model = YOLO(model_path)
    model.to(device) # Move model to the correct device
    model.eval() # Set the model to evaluation mode
    print(f"Model loaded successfully from {model_path} and moved to {device}.")
    # You can now use 'model' for inference
except Exception as e:
    print(f"Error loading model: {e}")
    print("Please ensure the file is a valid YOLO model and compatible with the current environment.")
```

### Perform Object Detection on a Sample Image

Now that the model is loaded, let's perform object detection on a sample image. We'll generate a dummy image for demonstration purposes.

```python
### Instructions for Local Path
1. Click the **Files** icon (folder shape) on the left sidebar.
2. Find your image file.
3. Right-click the file and select **'Copy path'**.
4. Paste that path into the `image_path` variable in the cell below.
```

Now, let's run the YOLO model on this dummy image.

```python
from PIL import Image
import matplotlib.pyplot as plt

# Paste your specific image path here
image_path = "/content/istockphoto-874779578-2048x2048.jpg"

# Run detection only on the specified image
results = model(image_path)

# Display only these specific results
for r in results:
    im_array = r.plot()  # plot predictions
    im = Image.fromarray(im_array[..., ::-1])  # Convert BGR to RGB for plotting
    plt.figure(figsize=(12, 8))
    plt.imshow(im)
    plt.axis("off")
    plt.show()
```
